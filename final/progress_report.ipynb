```python
import utils
import operator
import csv
```

```python
texts = []
with open('./corpus-abstracts.csv') as f:
    csvreader = csv.reader(f)
    texts += [row[0] for row in csvreader]
```

```python
data = utils.load_obj('./tfidf-chunks')
```

```python
# Get TF-IDF ranking over all chunks in the corpus
corpus = data['corpus']
dictionary = data['dictionary']
tfidf = data['tfidf']
corpus_tfidf = tfidf[corpus]


tfidf_by_word = {}
for doc in corpus_tfidf:
    for id, value in doc:
        word = dictionary.get(id)
        tfidf_by_word[word] = value

sorted_tfidf = sorted(tfidf_by_word.items(), key=operator.itemgetter(1))
sorted_tfidf.reverse()

sorted_tfidf
```

```python
# Get TF-IDF ranking over just the first abstract
abstract_tfidf = {}
for id, value in corpus_tfidf[0]:
    word = dictionary.get(id)
    abstract_tfidf[word] = value
sorted_abstract_tfidf = sorted(abstract_tfidf.items(), key=operator.itemgetter(1))
sorted_abstract_tfidf.reverse()

sorted_abstract_tfidf
```
